Tome as retas paralelas $r_1(t)=a_1+b_1t$ e $r_2(t)=a_2+b_2t$, onde $a_1,a_2,b_1,b_2\in\mathbb{R}^2$ e $t\in\mathbb{R}$. 

Se $r_1$ e $r_2$ são paralelas, então $b_1$ é múltiplo de $b_2$. Assim, $b_1=kb_2$ para algum $k\in\mathbb{R}$. Se $r_1$ e $r_2$ são coincidentes, então $a_1=a_2$ e $b_1=kb_2$ para algum $k\in\mathbb{R}$.

Seja $T:\mathbb{R}^2\to\mathbb{R}^2$ uma transformação linear definida por $T(x)=Ax+b$, onde $A\in\mathbb{R}^{2\times 2}$ e $b\in\mathbb{R}^2.$

Aplicando a transformação $T$ em $r_1$ e $r_2$, temos que:

\begin{aligned}
T(r_1)=Ar_1+b = A(a_1+b_1t)+b = \\
Aa_1+b + (Ab_1)t = Aa_1+b + (Akb_2)t = \\

Aa_1 + b + (Ab_2)kt = r_1'
\end{aligned}

$$T(r_2)=Ar_2+b = A(a_2+b_2t)+b = Aa_2+b + (Ab_2)t = (Aa_2+b) + (Ab_2)t = r_2'$$

Assim, $r_1'$ e $r_2'$ são retas paralelas pois $(Ab_2)kt$ é múltiplo de $(Ab_2)t$.

**Verdadeira**

Dados dois espaços métricos $X$ e $Y$, o mapa $f:X\rightarrow Y$ é isomeria se preserva a distância entre $a,b\in X$, ou seja, 

$d_X(a,b)=d_Y(f(a),f(b))$.

Assim, no contexto euclidiano, uma isomeria que leva a origem na origem é tal que $f(0)=0$ e :

$$||f(a)-f(0)|| = ||a-0||$$

$$||f(a)|| = ||a||$$

Logo, considerando $a$ e $b$ vetores e seja $\theta$ o ângulo entre eles, de modo que:

$$\cos{\theta}= \frac{a\cdot b}{||a|| ||b||}$$

Considere o ângulo entre $f(a)$ e $f(b)$, de modo que:

$$\cos{\theta'}= \frac{f(a)\cdot f(b)}{||f(a)|| ||f(b)||}$$

Veja que, pela isometria, temos que $||f(a)|| = ||a||$ e $||f(b)|| = ||b||$, e além disso:

$$||a-b|| = ||f(a)-f(b)||
$$

$$||a-b||^2 = ||f(a)-f(b)||^2$$

$$ (a-b)^T(a-b) = (f(a) - f(b))^T (f(a)-f(b)) $$

$$ a^Ta - 2a^Tb + b^Tb = f(a)^Tf(a) - 2f(a)^Tf(b) + f(b)^Tf(b)$$

Disso, temos que:

$$a^Tb = f(a)^Tf(b) $$

ou seja, 

$$a\cdot b = f(a)\cdot f(b)$$

Portanto, $\cos{\theta} = \cos{\theta'}$ e, assim, a isometria preserva o ângulo entre vetores.

Demonstração:

Se quatro pontos no plano projetivo $P'=\begin{bmatrix}
p'_1 & p'_2 & p'_3 & p'_4
\end{bmatrix}$ formam um quadrilatero não degenerado, então esses pontos definem uma base projetiva. Uma base projetiva em $\mathbb{R}\mathbb{P}^n$ é um conjunto de $n+2$ pontos tal que qualquer subconjunto de $n+1$ pontos é linearmente independente (como um conjunto de vetores em $\mathbb{R}^{n+1}$).

Assim, como a base canônica de $\mathbb{R}P^2$ é dada por 

$e_1=(1,0,0)$ 

 $e_2=(0,1,0)$ 
 
 $e_3=(0,0,1)$ 

 $e_4=(1,1,1)$=$e_1+e_2+e_3$.   

**Colorário 1**

**Dada qualquer base projetiva $a_1,a_2,a_3$, existe uma transformação projetiva $H$ tal que $H(e_i)= \lambda_i a_i$ para $i=1,2,3,4$.**

Prova:

$e_1,e_2,e_3$ são indenpendentes, logo as equações 
\begin{aligned}
H(e_1)=\lambda_1 a_1 \\
H(e_2)=\lambda_2 a_2 \\
H(e_3)=\lambda_3 a_3\\
\end{aligned}

definem uma transformação linear $H:\mathbb{R}^{n+1}\to\mathbb{R}^{n+1}$ cuja matriz $M$ de transformação, nas bases $e_1,e_2,e_3$ e $a_1, a_2, a_3$ é dada por:

$$M=
\begin{bmatrix}
\lambda_1 a_1 & \lambda_2 a_2 & \lambda_3 a_3
\end{bmatrix}
$$

Logo, a igualdade $H(e_4)=\lambda_4 a_4$ é dada na forma matricial por:

$$
\begin{bmatrix}
\lambda_1 a_1 & \lambda_2 a_2 & \lambda_3 a_3
\end{bmatrix}
\begin{bmatrix}
1 \\
1 \\
1
\end{bmatrix}
=
\lambda_4 a_4
$$

$$
\begin{bmatrix}
a_1 & a_2 & a_3
\end{bmatrix}
\begin{bmatrix}
\lambda_1 \\
\lambda_2 \\
\lambda_3
\end{bmatrix}
=
\lambda_4 a_4
$$

Assim, essa equação define unicamente $\lambda_1, \lambda_2, \lambda_3$.  Logo, a transformmação projetiva $H$ está definida, de acordo com a escolha de $\lambda_4$.

Assim, além da base canônica, esse rersultado se estende facilmente para qualquer base projetiva $b_1,b_2,b_3,b_4$.

Dessa forma, dadas duas bases projejtivas $a_1, a_2, a_3, a_4$ e $b_1, b_2, b_3, b_4$, existe uma única transformação projetiva $H$ tal que $H(a_i)=b_i$ para $i=1,2,3,4$, a menoss de uma constante de escala.

Para transformar um quadrilátero $Q_1$ em um quadrilátero $Q_2$, basta aplicar uma transformação projetiva $H_1:I^2\to Q_1$ que leva um quadrado unitário $I^2=[0,1]\times[0,1]$ em $Q_1$ e uma transformação projetiva $H_2:I^2\to Q_2$ que leva um quadrado unitário $I^2$ em $Q_2$. Assim, a transformação projetiva $H=H_2\circ H_1^{-1}$ leva $Q_1$ em $Q_2$.

Agora, ilustrando a transformação $T_1$, que levam os vértices de $I^2$ em $Q_1$:

Sejam, conforme no enunciado, $e_1=p_1=(0,0,1)$, $e_2=p_2=(0,1,1)$, $e_3=p_3=(1,1,1)$ e $e_4=p_4=(1,0,1)$.

Sejam $q_1=p_1'=(x_1,y_1,1)$, $q_2=p_2'=(x_2,y_2,1)$, $q_3=p_3'=(x_3,y_3,1)$ e $q_4=p_4'=(x_4,y_4,1)$ as coordenadas projetivas dos vértices do quadrilátero $Q_1$.

Assim:

$$
\begin{aligned}
H(e_1)=q_1 = (\lambda_1 x_1, \lambda_1 y_1, \lambda_1)
\\
H(e_2)=q_2 = (\lambda_2 x_2, \lambda_2 y_2, \lambda_2)
\\
H(e_3)=q_3 = (\lambda_3 x_3, \lambda_3 y_3, \lambda_3)
\end{aligned}
$$

Logo, a matriz de transformação $M$ é dada por:

$$
M=
\begin{bmatrix}
\lambda_1 x_1 & \lambda_2 x_2 & \lambda_3 x_3 \\
\lambda_1 y_1 & \lambda_2 y_2 & \lambda_3 y_3 \\
\lambda_1 & \lambda_2 & \lambda_3
\end{bmatrix}
=
\begin{bmatrix}
x_1 & x_2 & x_3 \\
y_1 & y_2 & y_3 \\
1 & 1 & 1
\end{bmatrix}
\begin{bmatrix}
\lambda_1 \\
\lambda_2 \\
\lambda_3
\end{bmatrix}
$$

Assim, como $H(e_4)=\lambda_4 q_4$, temos que:


$$
\begin{bmatrix}
x_1 & x_2 & x_3 \\
y_1 & y_2 & y_3 \\
1 & 1 & 1
\end{bmatrix}
\begin{bmatrix}
\lambda_1 \\
\lambda_2 \\
\lambda_3
\end{bmatrix}
=
\lambda_4 
\begin{bmatrix}
x_4 \\
y_4 \\
1
\end{bmatrix}
$$

Ffixando $\lambda_4=1$ esse sistema admitee euma solução única para $\lambda_1, \lambda_2, \lambda_3$ e substituindo na matriz $M$ temos a matriz de transformação $M$.

$p_i=\begin{bmatrix}

Vamos provar que existe euma única homografiai (transformação projetiva) $H:\mathbb{R}\mathbb{P}^2\to\mathbb{R}\mathbb{P}^2$ tal que $p'_i=Hp_i$.

Suponha por absurdo que existam duas homografias $H_1$ e $H_2$ tais que $P'=H_1P$ e $P'=H_2P$.

Uma homografia é simétrica, de fomra que combinando as duas equações, temos que:

$$H_1P=H_2P$$

$$H_2^{-1}H_1P=P$$
, donde concluímos que $H_2^{-1}H_1=I$, pela unicidadee da inversa, e então $H_1=H_2$.
